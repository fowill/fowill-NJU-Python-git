```python
### Version1 ###

#该程序的目的：在Bing上搜索我们想要的汽车公司+姓氏/职位等关键词，把符合条件的搜索结果网址存储下来
#请保证配置好了chromedriver和selenium

from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import MoveTargetOutOfBoundsException
from selenium.webdriver.chrome.options import Options
import time
import os

#自定义一个函数，让程序可以根据我们设定的最大等待时间，等待一个按钮出现并点击它（防止网页加载较慢，我们要点击的元素没有出现）
def click_by_time(driver,xpath,maxTime):
    t=0
    while t<=maxTime:
        if driver.find_element_by_xpath(xpath)!=None:  #如果找到了就点击它并退出循环，不然再等一秒
            driver.find_element_by_xpath(xpath).click()
            break
        time.sleep(1)
        t+=1  

surname_list = ['王','李','张','刘','陈']
occupation_list = ['Manager','Engineer','Purchasing','Logistics','R&D','Vice','工程','技术','经理','供应','物流']
big_companys = ['BYD Motor','比亚迪','奇瑞','大众','特斯拉']
eng_companys = ['BYD Motor']


for i in range(len(big_companys)): #对我们要搜素的公司循环
    company_urls = [] #company_urls列表存储某个公司所有的符合条件的搜索结果
    addition_list = surname_list + occupation_list #addition_list用来存储需要添加的关键词

    if big_companys[i] in eng_companys:
        addition_list = occupation_list #如果是英文公司名，addition_list就不要搜中国姓氏了

    for adKey in addition_list: 
        
        #拼凑出搜索关键词，site:linkedin.com指限定搜索结果在linkedin站点内
        key = adKey + ' ' + big_companys[i] + ' ' + ' site:linkedin.com' 

        #这里用了相对路径，我把chromedriver文件放到了当前文件夹的父目录（../表示）下的chromedriver文件夹中
        chrome_driver = '../chromedriver/chromedriver' #更换成各位电脑里chromedriver文件的路径
        
        driver = webdriver.Chrome(chrome_driver) #新建一个Chrome浏览器对象
        
        #step1 操纵浏览器打开bing
        driver.get('https://cn.bing.com/')
        
        #step2 在Bing搜索框里搜索我们需要的关键字，并点击搜索按钮
        searchingBox = driver.find_element_by_xpath('//*[@id="sb_form_q"]') #定位到搜索框元素
        searchingBox.send_keys(key) #在搜索框输入
        click_by_time(driver,'//*[@id="sb_form_go"]',10) #定位并点击搜索按钮，这函数是开头我们自定义的
         
        #step3 定位每一页的搜索结果，在其中辨别我们需要的结果
        count = 0
        records = 0
        while(True):
            if count == 100: #到100页就退出
                break
            count += 1
            time.sleep(3)#睡一下，防止未加载，是空列表
            
            titles = driver.find_elements_by_class_name('b_algo') #通过classname属性定位到所有搜索结果
            for title in titles:
                result = title.find_element_by_tag_name('a') #通过tag_name属性进一步定位到搜索结果标题
                if ('|' in result.text and '-' in result.text): #我们通过标题中是否有|等来识别是否是我们要的结果，这里不做解释
                    spans = title.find_elements_by_tag_name('a')
                    for span in spans:
                        href = span.get_attribute('href')
                        if 'www.linkedin' in str(href) or 'cn.linkedin' in str(href): #只要领英国际版和领英中国版的网址，美国版等不要
                            print(str(href))
                            path = os.getcwd() #获取当前工作路径
                            company_urls.append(str(href)) #把这次抓取到的信息添加到company_urls列表中
                            
                            #把当前抓取到的信息追加存储到本地，路径为当前工作路径下，名为xxxurls.txt的文件
                            with open(path + '/' + big_companys[i].replace(' ','') + 'urls.txt','a+') as f: #a+表示添加
                                f.write(str(href))
                                f.write('\n')
                            records += 1
                        break
            time.sleep(1)
            company_urls = list(set(company_urls)) #company_urls中的内容常常有重复，转化成set集合后可以去重，再转回list列表
            print(count,'页','第',i,'家企业',big_companys[i],'目前找到',len(company_urls),'条有效记录,辅助关键词为 ',adKey)
        
        #step4 点击“下一页”按钮，直到达到100页上限或者最后一页
            pagebar = driver.find_element_by_class_name('b_pag') 
            next_page = pagebar.find_elements_by_tag_name('a')[-1]
            next_page.click()
```
